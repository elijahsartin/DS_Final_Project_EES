```python
from bs4 import BeautifulSoup
```

```python
#creates object in BS containing html from site
URL = "https://www.basketball-reference.com/boxscores/202301230DET.html"
page = requests.get(URL, allow_redirects=False)

soup = BeautifulSoup(page.content, "html.parser")
```

```python
datetime = soup.findAll("div", {"class": "scorebox_meta"})
print(datetime)
```

```python
teams = soup.findAll({'strong': {'class': 'scorebox'}}, limit = 1)
print(teams)
```
